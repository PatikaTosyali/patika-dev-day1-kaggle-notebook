## 3. GitHub Repo Nedir?

### Neden?
Çünkü “repo” kelimesini duymadan GitHub'ı anlamak zordur.

### Ne?
**Repository (repo)**, bir projenin düzenli şekilde tutulduğu klasördür.

İçinde genelde şunlar olur:
- `README.md`
- `requirements.txt`
- `src/`
- `notebooks/`
- `data/`
- `models/`

### Benzetme
Bir repo, **düzenli bir proje dolabı** gibidir.

- README = dolabın kapağındaki açıklama etiketi
- requirements = bu dolabı kullanmak için gereken araç listesi
- src = asıl çalışma araçları
- notebooks = deneme / anlatım alanı
- data = malzeme
- models = üretilen modeller

### ASCII Şema

```text
project-repo/
├── README.md
├── requirements.txt
├── notebook.ipynb
├── src/
├── data/
└── models/
```

### README ne işe yarar?
Projenin “bu proje ne yapıyor ve nasıl başlatılır?” cevabıdır.

### requirements.txt ne işe yarar?
Projenin ihtiyaç duyduğu Python paketlerinin listesidir.

### Bir projeyi ilk gördüğümüzde nereye bakarız?
1. README
2. requirements
3. notebook veya src klasörü

- Bir projeyi daha az korkutucu hale getirir

## 6. GitHub Üzerinden Paket Kurma Mantığı

### Neden?
Bazen sadece bir repoyu klonlamak yetmez. O repo bazı paketler ister.

### Ne?
İki temel yaklaşım vardır:

1. `!pip install -r requirements.txt`
2. `!pip install git+https://github.com/...`

### Aralarındaki fark
- `requirements.txt` = proje sahibinin hazırladığı paket listesi
- `git+https://...` = bir paketi doğrudan GitHub reposundan kurmak

### Benzetme
- requirements = yemek tarifi için malzeme listesi
- git+ install = malzemenin doğrudan üreticisinden almak


### Örnek 1: requirements dosyasından kurulum
Bu hücre, gerçek kurulum komutunu gösterir. Eğitim sırasında isterseniz çalıştırabilirsiniz.

### Kod


````python
RUN_REQUIREMENTS_INSTALL = False

if RUN_REQUIREMENTS_INSTALL:
    !pip install -q -r requirements.txt
    print("Kurulum tamamlandı.")
else:
    print("Demo güvenlik için kapalı. Çalıştırmak isterseniz RUN_REQUIREMENTS_INSTALL = True yapın.")
````

### Örnek 2: Paketi doğrudan GitHub'dan kurmak
Bu örnek mantığı öğretmek içindir. Çok sık değil, ama bazen gerekir.

### Kod


````python
RUN_GITHUB_INSTALL_DEMO = False

if RUN_GITHUB_INSTALL_DEMO:
    !pip install -q git+https://github.com/psf/requests.git
    import requests
    print("requests sürümü:", requests.__version__)
else:
    print("GitHub'dan kurulum demosu kapalı. Gerçek derste ihtiyaç olursa açılabilir.")
````

### Kurulum sonrası küçük import testi
Çoğu zaman kurulumdan sonra en basit kontrol, import etmektir.

### Kod


````python
modules_to_try = ["numpy", "matplotlib", "cv2"]

for module_name in modules_to_try:
    try:
        __import__(module_name)
        print(f"[OK] {module_name} import edildi")
    except Exception as e:
        print(f"[HATA] {module_name} import edilemedi -> {e}")
````

### Bu yöntem ne zaman işimize yarar?
- Projede requirements dosyası varsa
- Public bir GitHub reposundaki paketi denemek istiyorsak
- Kaggle ortamında eksik paket varsa

### Kurulum başarısız olursa neyi kontrol ederiz?
- İnternet açık mı?
- requirements dosyası doğru klasörde mi?
- Repo içine girdik mi?
- Paket adı doğru mu?

### Mini Özet
`requirements.txt` proje ortamını topluca kurar. `git+https://...` ise özel durumlarda doğrudan GitHub kaynağından kurulum yapar.

| Yanlış dizinde komut çalıştırmak | Nerede olduğunuzu kontrol etmediniz | Önce `!pwd` veya `os.getcwd()` ile konumu görün |
| Clone sonrası repo içine girmemek | `%cd repo_adi` adımı atlandı | Repo klasörüne geçin |
| requirements kurulmadan kod çalıştırmak | Paketler hazır değil | Gerekirse `pip install -r requirements.txt` çalıştırın |
| Kaggle dosya yollarını karıştırmak | `input` ve `working` ayrımı unutuldu | Veri için `/kaggle/input`, çıktı için `/kaggle/working` düşünün |
| İnternet kapalıyken clone denemek | Notebook internet erişimi kapalı | Kaggle notebook ayarlarından internet durumunu kontrol edin |
